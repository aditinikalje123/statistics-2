4)What are probability distribution functions (PDF)
Ans:-In probability theory, a Probability Distribution Function (PDF) refers to a function that describes the likelihood of a random variable taking on a particular value or range of values.

5)How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)
Ans:-A Cumulative Distribution Function (CDF) gives the probability that a random variable is less than or equal to a certain value, while a Probability Distribution Function (PDF) (for continuous variables) or Probability Mass Function (PMF) (for discrete variables) describes the likelihood of specific values. The PDF shows how dense the probabilities are around a value, but the CDF accumulates those probabilities up to a point. For example, while a PDF tells you how likely a value is, the CDF tells you how likely it is to get a value up to that point. CDFs always increase from 0 to 1.

17)What are point estimates and interval estimates in statistics
Ans:-A point estimate is a single value calculated from sample data that serves as the best guess or estimate of an unknown population parameter (like the mean or proportion).
An interval estimate provides a range of values within which the population parameter is likely to lie, along with a confidence level.

18)What is the significance of confidence intervals in statistical analysis
Ans:-Confidence intervals (CIs) are crucial in statistics because they provide a range of plausible values for an unknown population parameter, rather than just a single point estimate. This helps quantify the uncertainty inherent in sampling.

19)What is the relationship between a Z-score and a confidence interval
Ans:-The Z-score and confidence interval (CI) are closely connected because the Z-score helps determine the width of the confidence interval when estimating a population parameter.

20)How are Z-scores used to compare different distributions
Ans:-Z-scores allow you to standardize values from different distributions so you can compare them on the same scale, even if the original data have different means and standard deviations.
